```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# URL de los ingresos de Tesla
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Usar un controlador de navegador (ejemplo con Chrome)
driver = webdriver.Chrome(executable_path='/path/to/chromedriver')  # Asegúrate de tener el chromedriver correcto
driver.get(url)

# Obtener el contenido HTML de la página después de que se haya cargado
html = driver.page_source

# Parsear el contenido con BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Encontrar la tabla de ingresos
table = soup.find('table', {'class': 'table'})

# Extraer las filas de la tabla
rows = table.find_all('tr')

# Crear una lista para almacenar los datos
data = []

# Recorrer las filas y extraer las celdas
for row in rows[1:]:  # Salto la primera fila (cabecera)
    cells = row.find_all('td')
    if len(cells) > 1:
        year = cells[0].get_text(strip=True)
        revenue = cells[1].get_text(strip=True)
        data.append([year, revenue])

# Crear un DataFrame con los datos extraídos
tesla_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Mostrar las últimas 5 filas
print(tesla_revenue.tail())

# Cerrar el navegador después de extraer los datos
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de los ingresos de GameStop (esto puede cambiar según la página)
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Hacer la solicitud a la página
response = requests.get(url)

# Parsear el contenido HTML de la página
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar la tabla de ingresos
table = soup.find('table', {'class': 'table'})

# Extraer las filas de la tabla
rows = table.find_all('tr')

# Crear una lista para almacenar los datos
data = []

# Recorrer las filas y extraer las celdas
for row in rows[1:]:  # Salto la primera fila (cabecera)
    cells = row.find_all('td')
    if len(cells) > 1:
        year = cells[0].get_text(strip=True)
        revenue = cells[1].get_text(strip=True)
        data.append([year, revenue])

# Crear un DataFrame con los datos extraídos
gme_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Mostrar las últimas 5 filas
print(gme_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar los datos históricos de Tesla (TSLA)
tesla_data = yf.download('TSLA', start='2020-01-01', end='2025-01-01')

# Crear la función make_graph para graficar el precio de las acciones
def make_graph(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Precio de Cierre de Tesla')
    plt.title('Gráfico de las Acciones de Tesla')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Llamar a la función para graficar los datos de Tesla
make_graph(tesla_data)
```
